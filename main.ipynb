```python
import requests
import urllib

query = """
[out:json][timeout:50];
area["admin_level"="6"]["name"="Monroe County"]["boundary"="administrative"]->.monroeCounty;

relation(area.monroeCounty)["admin_level"="6"]["boundary"="administrative"];
out geom;

way(area.monroeCounty)["highway"];
out geom;
"""

encoded = urllib.parse.quote(query)
url = "https://overpass-api.de/api/interpreter?data=" + encoded

response = requests.get(url)

if response.status_code == 200:
    with open('monroe-county.json', 'w') as file:
        file.write(response.text)
else:
    print("Error: " + str(response.status_code))
```

```python
import json

def convert(json_file, output_file):
    with open(json_file, 'r') as file:
        data = json_file.load(file)
    geojson = {
        "type": "FeatureCollection",
        "features": []
    }
    
    for element in data['elements']:
        feature = {
            "type": "Feature",
            "properties": {
                "id": element.get('id'),
                "tags": element.get('tags', {})
            },
            "geometry": None
        }
        if element['type'] == 'way' and 'highway' in element.get('tags', {}):
            feature['geometry'] = {
                "type" : "LineString",
                "coordinates": [[point['lon'], point['lat']] for point in element['geometry']]
            }
            
        elif element['type'] == 'relation' and 'boundary' in element.get('tags', {}):
            if element.get('tags', {}).get('type') == 'multipolygon':
                outer_coords = []
                inner_coords = []
                
                for member in element.get('members', []):
                    if member['role'] == 'outer':
                        outer_coords.append([[point['lon'], point['lat']] for point in member.get('geometry', [])])
                    elif member['role'] == 'inner':
                        inner_coords.append([[point['lon'], point['lat']] for point in member.get('geometry', [])])
                
                feature['geometry'] = {
                    "type": "MultiPolygon",
                    "coordinates": [outer_coords, inner_coords]
                }
        if feature['geometry']:
            geojson['features'].append(feature)
                
        with open(output_file, 'w') as file:
            json.dump(geojson, file)
```
